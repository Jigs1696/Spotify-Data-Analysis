```python
most_popular = df.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df.set_index("release_date", inplace = True)
df.index = pd.to_datetime(df.index)
df.head()
```

```python
df[["artists"]].iloc[18]
```

```python
df["duration"] = df["duration_ms"].apply(lambda x: round(x/1000))
```

```python
df.drop("duration_ms" , inplace = True , axis=1)
```

```python
df.duration.head()
```

```python
corr_df = df.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize = (15,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,cmap="Accent",linewidths=1,)
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
sample_df = df.sample(int(0.004*len(df)))
print(len(sample_df))
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df,y="loudness",x="energy",color = "c").set(title="Loudness Vs Energy Correletion")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df,y="popularity",x="acousticness",color = "c").set(title="Popularity Vs Acousticness Correletion")
```

```python
df['date']=df.index.get_level_values('release_date')
df.dates = pd.to_datetime(df.date)
years = df.dates.dt.year
```

```python
sns.displot(years,discrete = True,aspect=2,height=5,kind="hist").set(title = "Number of songs per year")
```

```python
total_dr = df.duration
fig_dims = (18,7)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years,y=total_dr,ax=ax,errwidth=False).set(title = "years Vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr = df.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years,y=total_dr,ax=ax).set(title = "years Vs Duration")
plt.xticks(rotation=60)
```
